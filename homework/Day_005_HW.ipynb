### 1. Create Dataframe

```python
df = pd.DataFrame(app_train)
df
```

### 2. 變量分析

```python
a = df['HOUR_APPR_PROCESS_START'].count()
b = df['HOUR_APPR_PROCESS_START'].sum()
c = df['HOUR_APPR_PROCESS_START'].mean()
d = df['HOUR_APPR_PROCESS_START'].min()
e = df['HOUR_APPR_PROCESS_START'].max()
f = df['HOUR_APPR_PROCESS_START'].median()
g = df['HOUR_APPR_PROCESS_START'].var()
h = df['HOUR_APPR_PROCESS_START'].std()
i = df['HOUR_APPR_PROCESS_START'].cumsum()
j = df['HOUR_APPR_PROCESS_START'].skew()
k = df['HOUR_APPR_PROCESS_START'].kurt()
print("About the 'HOUR_APPR_PROCESS_START' column")
print("非NA數: %s" %a)
print("總和: %s" %b)
print("平均: %s" %c)
print("最小值: %s" %d)
print("最大值: %s" %e)
print("中位數: %s" %f)
print("變異數: %s" %g)
print("標準差: %s" %h)
print("偏度:%s" %j)
print("峰度:%s" %k)
print("累加值:\n%s" %i)
df['HOUR_APPR_PROCESS_START'].describe()
```

### 3. 直方圖(Histogram)

```python
data = df['HOUR_APPR_PROCESS_START']
plt.hist(data, bins=24, color = 'steelblue', edgecolor = 'k')
plt.show()
```
